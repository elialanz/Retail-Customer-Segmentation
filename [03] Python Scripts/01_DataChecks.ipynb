```python
print("=" * 60)
print("ORDERS.CSV")
print("=" * 60)
df_ords = pd.read_csv(os.path.join(path, '[2] Data/[1] Original Data', 'orders.csv'))
print("Shape:", df_ords.shape)
print("\nDtypes:")
print(df_ords.dtypes)
print("\nNulls:")
print(df_ords.isnull().sum())
print("\nDuplicates:", df_ords.duplicated().sum())
print("\nSample:")
print(df_ords.head(5).to_string())
print("\ndescribe:")
print(df_ords.describe())

print("\n" + "=" * 60)
print("ORDERS_PRODUCTS_PRIOR.CSV")
print("=" * 60)
df_prior = pd.read_csv(os.path.join(path, '[2] Data/[1] Original Data', 'orders_products_prior.csv'))
print("Shape:", df_prior.shape)
print("\nDtypes:")
print(df_prior.dtypes)
print("\nNulls:")
print(df_prior.isnull().sum())
print("\nDuplicates:", df_prior.duplicated().sum())
print("\nSample:")
print(df_prior.head(5).to_string())
print("\ndescribe:")
print(df_prior.describe())

print("\n" + "=" * 60)
print("MERGE KEY CHECK")
print("=" * 60)
print("order_id dtype in orders:", df_ords['order_id'].dtype)
print("order_id dtype in prior:", df_prior['order_id'].dtype)
orders_ids = set(df_ords['order_id'])
prior_ids = set(df_prior['order_id'])
print("order_ids in prior but NOT in orders:", len(prior_ids - orders_ids))
print("order_ids in orders but NOT in prior:", len(orders_ids - prior_ids))
```
